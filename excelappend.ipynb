```python
#original
content = '''
Sure, here's another example of plagiarized content:

Original content: The Declaration of Independence was signed on July 4, 1776, in Philadelphia, Pennsylvania, by 56 delegates to the Continental Congress.

Plagiarized content: In Philadelphia, Pennsylvania, on July 4, 1776, 56 delegates to the Continental Congress signed the Declaration of Independence.
'''#plagiarized
copied = '''
Climate change is a pressing global issue that is causing significant changes to the Earth's weather patterns, natural ecosystems, and human societies. The primary cause of climate change is the increase in atmospheric concentrations of greenhouse gases, such as carbon dioxide, methane, and nitrous oxide, due to human activities such as burning fossil fuels, deforestation, and agriculture. The consequences of climate change are already being felt, with rising sea levels, more frequent and severe natural disasters, and shifts in agricultural productivity. Urgent action is needed to mitigate the effects of climate change and adapt to the changes that are already occurring.
 '''# print(len(content))
```

```python
row = []
```

```python
#stop word count
original_stopword_count = clean_text_regex.stopwordcount(content)
plagiarized_stopword_count = clean_text_regex.stopwordcount(copied)
#append to df
row.append(original_stopword_count)
row.append(plagiarized_stopword_count)
```

```python
#punctuation count
original_punctuation_count = clean_text_regex.punctuation_count(content)
plagiarized_punctuation_count = clean_text_regex.punctuation_count(copied)
#append to df
row.append(original_punctuation_count)
row.append(plagiarized_punctuation_count)
```

```python
#text after punctuation removal, stop word removal, numbers removal and extra whitespace removal
content = clean_text_regex.clean_text(content)
copied = clean_text_regex.clean_text(copied)
# print(content)
```

```python
#separating original content from plagiarized content
original,plagiarized = content,copied
# print(original)
# print(plagiarized)
```

```python
#word overlap ratio
import wordoverlapratio
row.append(wordoverlapratio.word_overlap_ratio(original,plagiarized))
# wordoverlapratio.word_overlap_ratio('hell','wate')
# df
```

```python
#cosine similarity
import cosine
row.append(cosine.cosine_similarity(original,plagiarized))
```

```python
#jaccard similarity
import jaccard
row.append(jaccard.jaccard_similarity(original,plagiarized))
```

```python
#euclidean distance
import euclidean_distance
row.append(euclidean_distance.euclidean(original,plagiarized))
```

```python
#Levenshtein Distance
import levenshtein
row.append(levenshtein.levenshtein_distance(original,plagiarized))
```
