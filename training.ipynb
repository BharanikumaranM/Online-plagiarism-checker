```python
#import necessary packages for training
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import training data
df = pd.read_excel('train.xlsx')
```

```python
#Feature Scaling using standard scalar euclidean_distance and levenshtein(train data)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df[['euclidean_distance','levenshtein']] = sc.fit_transform(df[['euclidean_distance','levenshtein']])
```

```python
#train data split to train and test
df = pd.DataFrame(df)
X_train = df.iloc[:,:-1]
y_train = df.iloc[:,-1]
```

```python
#Import test data
test = pd.read_excel('test.xlsx')
test
```

```python
#Feature Scaling using standard scalar euclidean_distance and levenshtein(test data)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
test[['euclidean_distance','levenshtein']] = sc.fit_transform(test[['euclidean_distance','levenshtein']])
```

```python
X_test = test.iloc[:,:-1]   
y_test = test.iloc[:,-1]
```

```python
#lazy prediction
from lazypredict.Supervised import LazyRegressor
reg = LazyRegressor(verbose=0,ignore_warnings=True, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)
models
```

```python
#linear support vector regression
from sklearn.svm import SVR
regressor = SVR(kernel = 'linear')
regressor.fit(X_train, y_train)
#predicting the test set results
y_pred = regressor.predict(X_test)
#evaluating the model performance
from sklearn.metrics import r2_score
# r2_score(y_test, y_pred)
#accuracy score
regressor.score(X_test, y_test)
#accuracy
from sklearn.metrics import accuracy_score
# accuracy_score(y_test, y_pred)
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
#predicting the test set results
y_pred = regressor.predict(X_test)
#evaluating the model performance
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
#accuracy score
regressor.score(X_test, y_test)
```

```python
#LarsCV
from sklearn.linear_model import LarsCV
regressor = LarsCV()
regressor.fit(X_train, y_train)
#predicting the test set results
y_pred = regressor.predict(X_test)
#evaluating the model performance
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
#average of all the models
from sklearn.ensemble import VotingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LarsCV
from sklearn.svm import SVR
reg1 = LinearRegression()
reg2 = LarsCV()
reg3 = SVR(kernel = 'linear')
ereg = VotingRegressor([('lr', reg1), ('lars', reg2), ('svr', reg3)])
ereg = ereg.fit(X_train, y_train)
#predicting the test set results
y_pred = ereg.predict(X_test)
#evaluating the model performance
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
y_pred
```
